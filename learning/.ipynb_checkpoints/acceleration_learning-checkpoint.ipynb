```python
import numpy as np
import sys, os, h5py, re
import matplotlib.pyplot as plt
```

```python
import torch as th
from torch.optim import Adam
from network import AccPolicy, setParams
import matplotlib.pyplot as plt
from tqdm import tqdm
from learning_utils import prepare_acc_data

device = th.device('cuda' if th.cuda.is_available() else 'cpu')
```

```python
train_path = "../data/training/2-dof-no-velocity/training_DMP25_D8.hdf5"
test_path = "../data/testing/2-dof-no-velocity/validating_DMP25_D4.hdf5"
X, Y = prepare_acc_data(train_path)
```

```python
index = np.arange(len(X))
import random
random.shuffle(index)
X, Y = X[index], Y[index]
```

```python
index_ = int(0.95*len(X))
Xtrain, Ytrain = X[:index_], Y[:index_]
Xvali, Yvali = X[index_:], Y[index_:]
```

```python
Xtrain, Ytrain = th.from_numpy(Xtrain).float().to(device), th.from_numpy(Ytrain).float().to(device)
Xvali, Yvali = th.from_numpy(Xvali).float().to(device), th.from_numpy(Yvali).float().to(device)
```

```python
net = AccPolicy(8, 2).to(device)
optimizer = Adam(net.parameters(), lr = 1e-3)
MSE = th.nn.MSELoss()
```

```python
n_epochs = 150
batch_size = 64
val = []
train = []

for epoch in tqdm(range(n_epochs)):
    permutation = th.randperm(Xtrain.size()[0])
    train_loss = 0.
    vali_loss = 0.
    epo_length = len(list(range(0, Xtrain.size()[0], batch_size)))
    for i in range(0, Xtrain.size()[0], batch_size):
        optimizer.zero_grad()
        indices = permutation[i:i+batch_size]
        batch_x, batch_y = Xtrain[indices], Ytrain[indices]
        outputs = net(batch_x)
        loss = MSE(outputs, batch_y)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    with th.no_grad():
        net.eval()
        pred = net(Xvali)
        vali_loss = MSE(pred, Yvali)
        net.train()
    val.append(vali_loss)
    train.append(train_loss/epo_length)
```

```python
plt.plot(val, '-b')
plt.plot(train, '-r')
```

## Test

```python
f = h5py.File(test_path, 'r')
for key in f.keys():
    q = f[key]['q']
    t = f[key]['t']
    q = np.array(q)
    start = np.copy(q[0].reshape(1,-1))
    end = np.copy(q[-1].reshape(1,-1))
    q_curr = np.copy(start)
    qd_curr = np.copy(np.zeros([1,2]))
    traj = []
    for i, _ in enumerate(t):
        x_input = np.concatenate([start, end, q_curr, qd_curr], axis=1)
        with th.no_grad():
            x_input = th.from_numpy(x_input).to(device).float()
            out_acc = net(x_input)
        traj.append(np.copy(q_curr))
        qd_curr += out_acc.detach().cpu().numpy() * 0.004
        q_curr += qd_curr * 0.004

    traj = np.concatenate(traj)
    plt.figure()
    plt.grid()
    plt.plot(traj, '-r', label='predict')
    plt.plot(q, '-b', label='true')
    handles, labels = plt.gca().get_legend_handles_labels()
    labels, ids = np.unique(labels, return_index=True)
    handles = [handles[i] for i in ids]
    plt.legend(handles, labels, loc='best')
```
