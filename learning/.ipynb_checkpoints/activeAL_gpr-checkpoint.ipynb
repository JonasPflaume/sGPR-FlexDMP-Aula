```python
from algpr.gpr import GaussianProcessRegressor
from algpr.kernels import RBF
```

```python
sym_robot = symbolic_robot.symbolic_robot(robot_name='IRB910_2r', tool_mass=0.0,
                                                tool_frame=[0., 0., 0.3, 0., 0., 0.], load_casadi_fnc=False)
q_min, q_max = sym_robot.q_min, sym_robot.q_max
qd_min, qd_max = np.zeros(2), sym_robot.qd_max
```

```python
q_max
```

```python
# start with 50 random sample
q_his = []
X_init = []
Y_init = []
for _ in range(50):
    qi = [np.random.uniform(q_min[i], q_max[i]) for i in range(2)]
    qf = [np.random.uniform(q_min[i], q_max[i]) for i in range(2)]
    qdi = [i for i in [0., 0.]]
    qdf = [i for i in [0., 0.]]
    dt, t, q, qd, qdd, tau, weight = collect_data_with_dmp_weight(sym_robot, qi, qf, qdi, qdf, DMP_ORDER=25)
    q_his.append(q)
    X_init.append(np.array(qi + qf).reshape(1,-1))
    y_i = np.append(weight.flatten(), len(t))
    Y_init.append(y_i.reshape(1,-1))
X_init = np.concatenate(X_init)
Y_init = np.concatenate(Y_init)
```

```python
Xv, Yv, Trueth = prepare_data("../data/testing/2-dof-no-velocity/validating_DMP25_D4.hdf5")
```

```python
gpr = GaussianProcessRegressor(kernel=RBF(l=np.array([2.]*4), anisotropic=True))
gpr.fit(X_init, Y_init, call_hyper_opt=False)

pred = gpr.predict(Xv)
error_time, error_with_q, dyna_error = calc_loss(pred, Yv, Trueth, 25, return_dyn_loss=True,\
                                     tau_func=sym_robot.ck.get_inverse_dynamics_rnea(), tau_max=sym_robot.tau_max)
print("Time error: {0:3f}, Shape error: {1:3f}, Dyna error: {2:3f}".format(error_time, error_with_q, dyna_error))
```

### Active learning main loop

```python
# give a 1% duration, for the Aula will violate the constraints a little bit
lbx = [q_min[i]*.999 for i in range(2)] + [q_min[i]*.999 for i in range(2)]
ubx = [q_max[i]*.999 for i in range(2)] + [q_max[i]*.999 for i in range(2)]
```

```python
%%time
import time
time_loss_l = []
shape_loss_l = []
computation_cost_l = []
length_scale = np.array([2.2]*4)
training_len = 1000
for k in range(training_len):
    gpr = GaussianProcessRegressor(kernel=RBF(l=length_scale, anisotropic=True), noise_level=0.)
    if (k % 5 == 0 and k < 200) or (k % 10 == 0 and k > 200 and k < 700) or (k % 30 == 0 and k > 700):
        gpr.fit(X_init, Y_init, call_hyper_opt=True)
    else:
        gpr.fit(X_init, Y_init)
    length_scale = gpr.kernel.length_scale
    start_t = time.time()
    x_next = gpr.max_entropy_x(lbx, ubx)
    print(x_next)
    end_t = time.time()
    computation_cost_l.append(end_t - start_t)
    print("At ", k , " sampling")
    qi = [x_next[0,0], x_next[1,0]]
    qf = [x_next[2,0], x_next[3,0]]
    qdi = [i for i in [0., 0.]]
    qdf = [i for i in [0., 0.]]
    dt, t, q, qd, qdd, tau, weight = collect_data_with_dmp_weight(sym_robot, qi, qf, qdi, qdf)

    X_init = np.append(X_init, np.array(qi + qf).reshape(1,-1), axis=0)
    y_i = np.append(weight.flatten(), len(t))
    Y_init = np.append(Y_init, y_i.reshape(1,-1), axis=0)
    pred = gpr.predict(Xv)
    time_loss, shape_loss, dyna_loss = calc_loss(pred, Yv, Trueth, 25, return_dyn_loss=True,\
                                     tau_func=sym_robot.ck.get_inverse_dynamics_rnea(), tau_max=sym_robot.tau_max)
    time_loss_l.append(time_loss)
    shape_loss_l.append(shape_loss)
    print("Time Loss: ", time_loss,"Shape Loss: ", shape_loss, "Dyna Loss: ", dyna_loss)
```

```python
plt.figure()
plt.plot(time_loss_l)
plt.xlabel("Samples")
plt.ylabel("loss")
plt.yscale('log')
plt.title("time loss")
plt.grid()
plt.plot(np.arange(len(time_loss_l)), np.ones(len(time_loss_l)) * 0.0055)
plt.savefig("/home/jiayun/Desktop/time_dmp.jpg", dpi=200)
plt.figure()
plt.plot(shape_loss_l)
plt.xlabel("Samples")
plt.ylabel("loss")
plt.yscale('log')
plt.title("shape loss")
plt.plot(np.arange(len(time_loss_l)), np.ones(len(time_loss_l)) * 3.2)
plt.grid()
plt.savefig("/home/jiayun/Desktop/shape_dmp.jpg", dpi=200)
X_init.shape
```
