```python
DMP_ORDER = 25
X, Y, _ = prepare_data("../data/training/2-dof-no-velocity/training_DMP25_D7.hdf5")
```

```python
Y.shape
```

```python
def GP_regression_std(regressor, X):
    _, std = regressor.predict(X, return_var=True)
    query_idx = np.argmax(std)
    return query_idx, X[query_idx], np.max(std)
```

### training with the whole dataset

```python
from modAL.models import ActiveLearner
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
```

```python
from algpr.gpr import GaussianProcessRegressor
from algpr.kernels import RBF
```

```python
kernel = RBF(l=2)
gpr = GaussianProcessRegressor(kernel=kernel)
gpr.fit(X, Y)
```

```python
Xv, Yv, Trueth = prepare_data("../data/testing/2-dof-no-velocity/validating_DMP25_D4.hdf5")
```

```python
pred = gpr.predict(Xv)
error_time, error_with_q = calc_loss(pred, Yv, Trueth, DMP_ORDER)
print("time error: %.3f" % error_time, "shape error: %.3f" % error_with_q)
```

### active learning

```python
n_initial = 10
initial_idx = np.random.choice(range(len(X)), size=n_initial, replace=False)
X_training, Y_training = X[initial_idx], Y[initial_idx]

regressor = ActiveLearner(
    estimator=GaussianProcessRegressor(kernel=kernel),
    query_strategy=GP_regression_std,
    X_training=X_training.reshape(-1, 4), y_training=Y_training.reshape(-1, DMP_ORDER*2+1)
)
```

```python
pred = regressor.predict(Xv)
error_time, error_with_q = calc_loss(pred, Yv, Trueth, DMP_ORDER)
print("time error: %.3f" % error_time, "shape error: %.3f" % error_with_q)
```

### start active learning to reduce the std

```python
sym_robot = symbolic_robot.symbolic_robot(robot_name='IRB910_2r', tool_mass=0.0,
                                                tool_frame=[0., 0., 0.3, 0., 0., 0.],load_casadi_fnc=False)
q_min, q_max = sym_robot.q_min, sym_robot.q_max
qd_min, qd_max = np.zeros(2), sym_robot.qd_max

def sample_one_batch(batch_size):
    q1_i = np.random.uniform(q_min[0], q_max[0], [batch_size, 1])
    q2_i = np.random.uniform(q_min[1], q_max[1], [batch_size, 1])
    q1_f = np.random.uniform(q_min[0], q_max[0], [batch_size, 1])
    q2_f = np.random.uniform(q_min[1], q_max[1], [batch_size, 1])
    q = np.concatenate([q1_i, q2_i, q1_f, q2_f], axis=1)
    return q
```

```python
error_time = float("inf")
# var_bound = float("inf")

time_error_l = []
shape_error_l = []
for _ in range(400):
    query_idx, query_instance, max_std = regressor.query(X)

    regressor.teach(X[query_idx].reshape(1, 4), Y[query_idx].reshape(1,DMP_ORDER*2+1))
    pred = regressor.predict(Xv)
    error_time, error_with_q = calc_loss(pred, Yv, Trueth, DMP_ORDER)
    time_error_l.append(error_time)
    shape_error_l.append(error_with_q)
    print("time error: %.3f" % error_time, "shape error: %.3f" % error_with_q)
```
