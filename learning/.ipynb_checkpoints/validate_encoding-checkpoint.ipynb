```python
info = np.zeros([5, 3, 3]) # spline, knots, (dyna_loss, dyna_ratio, dyna_peak)
for i, spline in enumerate([CubicSpline, Order5Spline, QuinticSpline, Order7spline, Order8spline]):
    for SPL_KNOTS in range(15,30,5):
        print(i, SPL_KNOTS)
#         X, Y, _ = prepare_data_via("../data/training/2-dof-no-velocity/training_SPL40_D8.hdf5", spline, knots=SPL_KNOTS)
        Xv, Yv, Trueth, qd_trueth, qdd_trueth = \
                prepare_data_via("../data/validating_D4_N30_jerk_.hdf5", spline, knots=SPL_KNOTS, return_dyddy=True)
        X, Y, _ = prepare_data_via("../data/training_D8_N30_jerk_.hdf5", spline=spline, knots=SPL_KNOTS)
        gpr = GaussianProcessRegressor(kernel=RBF(l=np.array([2]*4), anisotropic=True))
        gpr.fit(X, Y)

        pred = gpr.predict(Xv)
        _, _, error_dyna, dyna_info = calc_loss_spl_via(pred, Xv, Yv, Trueth, spline, SPL_ORDER=SPL_KNOTS, \
                    return_dyn_loss=True, return_dyn_info=True, tau_func=sym_robot.ck.get_inverse_dynamics_rnea(), tau_max=sym_robot.tau_max)
        dyna_peak, dyna_ratio = dyna_info["vio_peak"], dyna_info["vio_rate"]
        info[i, SPL_KNOTS//5-3, 0] = error_dyna
        info[i, SPL_KNOTS//5-3, 1] = dyna_ratio
        info[i, SPL_KNOTS//5-3, 2] = np.max(dyna_peak)
```

#### Grid search

```python
info = np.zeros([5, 15, 3]) # spline, knots, (dyna_loss, dyna_ratio, dyna_peak)
for i, spline in enumerate([CubicSpline, Order5Spline, QuinticSpline, Order7spline, Order8spline]):
    for SPL_KNOTS in range(15,30):
        print(i, SPL_KNOTS)
#         X, Y, _ = prepare_data_via("../data/training/2-dof-no-velocity/training_SPL40_D8.hdf5", spline, knots=SPL_KNOTS)
        Xv, Yv, Trueth, qd_trueth, qdd_trueth = \
                prepare_data_via("../data/validating_D4_N30_jerk_.hdf5", spline, knots=SPL_KNOTS, return_dyddy=True)

        pred = Yv
        _, _, error_dyna, dyna_info = calc_loss_spl_via(pred, Xv, Yv, Trueth, spline, SPL_ORDER=SPL_KNOTS, \
                    return_dyn_loss=True, return_dyn_info=True, tau_func=sym_robot.ck.get_inverse_dynamics_rnea(), tau_max=sym_robot.tau_max)
        dyna_peak, dyna_ratio = dyna_info["vio_peak"], dyna_info["vio_rate"]
        info[i, SPL_KNOTS-15, 0] = error_dyna
        info[i, SPL_KNOTS-15, 1] = dyna_ratio
        info[i, SPL_KNOTS-15, 2] = np.max(dyna_peak)
```

```python
plt.figure(figsize=[7,5])
for i in range(5):
    plt.plot(info[i,:,0], label="{}th polynomial".format(str(i+4)))
plt.legend()
plt.xlabel("knots number")
plt.ylabel("loss")
plt.title("The dynamic error")
plt.grid()
plt.savefig('/home/jiayun/Desktop/dyna_error.jpg', dpi=150)

plt.figure(figsize=[7,5])
for i in range(5):
    plt.plot(info[i,:,1], label="{}th polynomial".format(str(i+4)))
plt.title("The violation percentage")
plt.legend()
plt.xlabel("knots number")
plt.ylabel("loss")
plt.grid()
plt.savefig('/home/jiayun/Desktop/percent.jpg', dpi=150)

plt.figure(figsize=[7,5])
for i in range(5):
    plt.plot(info[i,:,2], label="{}th polynomial".format(str(i+4)))
plt.title("The violation peak")
plt.xlabel("knots number")
plt.ylabel("loss")
plt.legend()
plt.grid()
plt.savefig('/home/jiayun/Desktop/peak.jpg', dpi=150)

# plt.figure(figsize=[7,5])
# for i in range(5):
#     plt.plot(info[i,:,0]/np.max(info[i,:,0])*0.1 + info[i,:,1]/np.max(info[i,:,1])*0.3 + info[i,:,2]/np.max(info[i,:,2])*0.6, label="{}th polynomial".format(str(i+4)))
# plt.title("Weighted score")
# plt.legend()
# plt.grid()
# print(np.argmin(info[4,:,0]/np.max(info[4,:,0])*0.1 + info[4,:,1]/np.max(info[4,:,1])*0.3 + info[4,:,2]/np.max(info[4,:,2])*0.6))
```

```python
gpr = GaussianProcessRegressor(kernel=RBF(l=np.array([1.9]*4), anisotropic=True))
gpr.fit(X, Y)
### large dataset, GPR unstable
```

```python
gpr = GpoeGPR(kernel=RBF(l=np.array([1.]*4), anisotropic=True), max_subset_num=1000)
gpr.fit(X, Y, style='local')
```

```python
gpr = PPGPR(kernel=RBF(l=np.array([1.9]*4), anisotropic=True))
gpr.fit(X, Y, m=2000)
```

```python
import time
time_l = []
for i in range(50):
    start = time.time()
    pred = gpr.predict(Xv)
    end = time.time()
    time_l.append((end - start))
time_l = np.array(time_l)
mean = time_l.mean()
var = ((time_l - mean) ** 2).mean()
print(mean, var)
```

```python
error_time, error_target_q, error_dyn_loss, dyna_info = calc_loss_flexdmp(pred, Xv, Yv, Trueth, DMP_ORDER, \
                 return_dyn_loss=True, return_dyn_info=True, tau_func=sym_robot.ck.get_inverse_dynamics_rnea(), tau_max=sym_robot.tau_max)
print(error_time, error_target_q, error_dyn_loss)
```
