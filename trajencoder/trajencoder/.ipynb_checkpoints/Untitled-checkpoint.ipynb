```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
FILE = h5py.File('/home/jiayun/MotionLearning/suboptimal_planner/data/6dof/new_dynamics/Vali_Box1ToBox2_4.0.hdf5')
random_index = list(FILE.keys())[1] # 129
y_demo, dy_demo, ddy_demo = np.array(FILE[random_index]['q']), np.array(FILE[random_index]['qd']), np.array(FILE[random_index]['qdd'])
```

```python
FILE.close()
```

```python
q = y_demo[:,0]
dq = dy_demo[:,0]
ddq = ddy_demo[:,0]
```

```python
ay = 25.
by = 6.25
goal = q[-1]
tau = 1.
```

```python
f_target = ddq - ay * (by * (goal - q) - dq)
```

```python
x = 1.
tau = 1.
bf_num = 20
xl = []
for _ in range(len(q)):
    xl.append(x)
    x += 0.004*(-1/tau * x)
xl = np.array(xl)

des_c = np.linspace(0, len(q)*0.004, bf_num)

c = np.ones(len(des_c))
for n in range(len(des_c)):
    # finding x for desired times t, here exp is the solution of the cs system!!!
    c[n] = np.exp(-1. * des_c[n])

h = np.ones(bf_num) * bf_num ** 2 / 0.1*c

class RBF:
    ''' RBF activation function
    '''
    def __init__(self, center, width):
        self.c = center
        self.h = width

    def __call__(self, x_):
        return np.exp(-self.h * (x_ - self.c) ** 2 )
    
act_l = [RBF(center, width) for center, width in zip(c, h)]

weight = np.zeros(bf_num)
for i in range(bf_num):
    S = xl * (q[-1] - q[0])
    Gam = np.diag( act_l[i](xl) )
    weight[i] = (S.T @ Gam @ f_target) / (S.T @ Gam @ S)
```

```python
q_ = []
dq_ = []
# ddq_ = []
q0 = q[0]
dq0 = dq[0]
# ddq0 = ddq[0]
x = 1.
tau = 1.
```
