```python
import numpy as np
from trajencoder.flexdmp.utils import RBF, integrate_RK4
import matplotlib.pyplot as plt
import h5py
import random
FILE = h5py.File('/home/jiayun/MotionLearning/suboptimal_planner/data/training/2-dof-no-velocity/training_SPL40_D4.hdf5', 'r')
random_index = list(FILE.keys())[129]
y_demo, dy_demo, ddy_demo = np.array(FILE[random_index]['q']), np.array(FILE[random_index]['qd']), np.array(FILE[random_index]['qdd'])

length = len(y_demo)
dt = 0.004
time = np.linspace(0, dt*length, length)
jointNO = 0
y_demo, dy_demo, ddy_demo = y_demo[:,jointNO], dy_demo[:, jointNO], ddy_demo[:, jointNO]
traj_time = dt * length
FILE.close()
```

```python
dddy_demo = np.gradient(ddy_demo)/dt
```

```python
Tend = traj_time
p = 10 / Tend
alpha_1 = 3 * p
alpha_2 = p
alpha_3 = 1/3 * p
tau = 1

initial = y_demo[0]
goal = y_demo[-1]
N_weight = 20
centers = np.linspace(1,0,N_weight)
Delta_c = np.abs(centers[1]-centers[0])
r= 0.5
stds = Delta_c/2 * np.sqrt(-1/(2*np.log(r))) * np.ones([N_weight, 1])
weights = np.zeros([N_weight, 1])
```

```python
s_vec = np.zeros([length,1])
f_target = np.zeros([length,1])
x_here = np.zeros([length,1])
x_here[0] = 1.
for i in range(0,length):
    x_here[i] = 1 - 1/(tau*Tend) * time[i]
    if x_here[i] < 0:
        x_here[i] = 0
        
    s_vec[i] = x_here[i] * (goal - initial)
    f_target[i] = alpha_1*alpha_2*alpha_3*(y_demo[i]-goal) + tau * alpha_1*alpha_2*dy_demo[i] + tau**2 * alpha_1*ddy_demo[i] + tau**3 * dddy_demo[i]
```

```python
for i in range(0, N_weight):
    Gamma_vec = np.zeros([1,length])
    for j in range(0, length):
        Gamma_vec[:,j] = np.exp(-1/(2*stds[i]**2) * (x_here[j] - centers[i])**2 )
    Gamma = np.diag(Gamma_vec.squeeze())
    weights[i] = s_vec.T @ Gamma @ f_target / (s_vec.T @ Gamma @ s_vec)
```

```python
def forcing_function(x, weights, centers, stds, goal, initial):
    stds = stds.squeeze()
    psi = np.exp(-1/(2*stds**2) * (x - centers) ** 2 ) # vec
    print(psi)
    return (np.dot(weights.squeeze(), psi) / psi.sum()) * x * (goal - initial)
```

```python
x = np.zeros([length,1])
x[0] = 1

y1_3 = np.zeros([length,1])
y2_3 = np.zeros([length,1])
y3_3 = np.zeros([length,1])
y1_3[0] = initial
acc_3 = np.zeros([length,1])
f_applied_3 = np.zeros([length,1])
for i in range(1,length):
    
    xdot = -1/(tau*Tend)
    x[i] = x[i-1] + dt * xdot
    if x[i] < 0:
        x[i] = 0
    
    f_applied_3[i-1] = forcing_function(x[i-1],weights,centers,stds,goal,initial)
    y1_3dot = 1/tau * y2_3[i-1]
    y2_3dot = 1/tau * y3_3[i-1] 
    y3_3dot = 1/tau * (alpha_1*(alpha_2 * (alpha_3*(goal-y1_3[i-1]) - y2_3[i-1]) - y3_3[i-1]) + f_applied_3[i-1])
    
    y1_3[i] = y1_3[i-1] + dt*y1_3dot
    y2_3[i] = y2_3[i-1] + dt*y2_3dot
    y3_3[i] = y3_3[i-1] + dt*y3_3dot 
    acc_3[i] = y2_3dot
```

```python
plt.plot(y1_3, '-.r')
plt.plot(y_demo)
plt.figure()
plt.plot(y2_3, '-.r')
plt.plot(dy_demo)
plt.figure()
plt.plot(y3_3, '-.r')
plt.plot(ddy_demo)
```
